```python
sando_results = requests.get(sando_url).json()
sando_results
```

```python
sando_top_picks = sando_results['response']['venues']

sando_df = json_normalize(sando_top_picks)
sando_df.head()
```

```python
sando_df.shape
```

```python
filtered_columns = ['name', 'categories']+[col for col in sando_df.columns if col.startswith('location.')] + ['id']
sando_df_filtered = sando_df.loc[:, filtered_columns]

def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
sando_df_filtered['categories'] = sando_df_filtered.apply(get_category_type, axis=1)
sando_df_filtered.columns = [column.split(".")[-1] for column in sando_df_filtered.columns] 

sando_df_filtered
```

```python
sf = sando_df_filtered.groupby('categories')
sf.head()
```

```python
sando_map = folium.Map(location=[sando_latitude, sando_longitude], zoom_start=16)

# add red circle to show starting point
folium.CircleMarker(
[sando_latitude, sando_longitude],
radius=10,
colour='red',
popup='San Fernando',
fill=True,
fill_color='red',
fill_opacity=0.6
).add_to(sando_map)

# add places in sando
for lat, lng, label in zip(sando_df_filtered.lat, sando_df_filtered.lng, sando_df_filtered.categories):
    folium.CircleMarker(
    [lat, lng],
    radius=3,
    color='blue',
    fill=True,
    fill_color='yellow',
    fill_opacity=0.6
    ).add_to(sando_map)
    
sando_map
```
